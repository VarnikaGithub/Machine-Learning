```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv(r"C:\Users\Hp\Downloads\archive (3)\Melbourne_housing_FULL.csv")
df.head(20)
```

```python
Columns_use=['Suburb','Rooms','Type','Price','Method','Distance','Regionname','Propertycount','Bedroom2','Bathroom','Car','Landsize',
            'BuildingArea','CouncilArea']
df=df[Columns_use]
```

```python
df.head(20)
```

```python
#Columns that we fill to zero
Columns_fill_zero=['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']
df[Columns_fill_zero]=df[Columns_fill_zero].fillna(0)
```

```python
#Columns that we fill with mean
df['Landsize'] = df['Landsize'].fillna(df.Landsize.mean())
df['BuildingArea'] = df['BuildingArea'].fillna(df.BuildingArea.mean())
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df.head()
```

```python
#One Hot Encoding
df = pd.get_dummies(df, drop_first=True)
```

```python
df.head()
```

```python
#Define X and Y variable
X=df.drop('Price',axis=1)
y=df['Price']
```

```python
#Test and train Dataset
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.3, random_state=2)
```

```python
#Fit a regression model
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(train_X, train_y)
```

```python
reg.score(test_X, test_y)
```

```python
reg.score(train_X, train_y)
```

```python
#We can clearly see that training data have overfitted
```

```python
#Fit Lasso Regression
from sklearn import linear_model
lasso_reg = linear_model.Lasso(alpha=50, max_iter=100, tol=0.1)
lasso_reg.fit(train_X, train_y)
```

```python
lasso_reg.score(test_X, test_y)
```

```python
lasso_reg.score(train_X, train_y)
```

```python
#Fit Ridge Resgression
from sklearn.linear_model import Ridge
ridge_reg= Ridge(alpha=50, max_iter=100, tol=0.1)
ridge_reg.fit(train_X, train_y)
```

```python
ridge_reg.score(test_X, test_y)
```

```python
ridge_reg.score(train_X, train_y)
```

```python
#We see that Lasso and Ridge Regularizations prove to be beneficial when our Simple Linear Regression Model overfits
```
